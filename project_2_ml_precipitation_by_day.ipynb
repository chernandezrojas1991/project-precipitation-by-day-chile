```python
import pandas as pd
import numpy as np
```

```python
filename = "/content/drive/MyDrive/Coding Dojo/Machine Learning/Proyecto 2/dataset.csv"
df_dirty = pd.read_csv(filename, header = None)
df_precipitaciones = pd.DataFrame(df_dirty.transpose().values)
```

```python
df_precipitaciones.columns = df_precipitaciones.iloc[0]
df_precipitaciones = df_precipitaciones[1:]
df_precipitaciones.reset_index(drop=True, inplace=True)
df_precipitaciones.head(10)
```

Limpiar dataframe para que solamente tenga las fechas desde el 1 enero del 2000 hasta fechas del 2020.

```python
regex = r'^19\d{2}-\d{2}-\d{2}$'
df_filtered = df_precipitaciones.loc[:, ~df_precipitaciones.columns.str.match(regex)]
```

Realizar transformación de columna respecto a las fechas registradas. Se crea una columna *fecha* y *mediciones*, con tal que la columna *fecha* tenga cada día registrado y *mediciones* tenga los milímetros de precipitaciones caídos ese día.

Nota: Un valor -9999 para *mediciones* significa que ese día no hubo lluvia.

```python
df_precip_clean = pd.melt(df_filtered, id_vars=['codigo_estacion', 'institucion', 'fuente', 'nombre', 'altura', 'latitud', 'longitud', 'codigo_cuenca', 'nombre_cuenca', 'codigo_sub_cuenca', 'nombre_sub_cuenca', 'inicio_observaciones', 'fin_observaciones', 'cantidad_observaciones', 'inicio_automatica'], var_name='fecha', value_name='mediciones')
df_precip_clean.head(10)
```
